## Preparing the Data

```python
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Tugas 4_Ekon Regional.csv', delimiter =",")

data.head()
```

```python
# Calculate total employment for each year
data['totalempsumsel'] = data.groupby('year')['employment_sumsel'].transform('sum')
data['totalempnational'] = data.groupby('year')['employment_national'].transform('sum')
```

```python
data.head()
```

## LQ, SLQ, and DLQ

```python
# Calculate Location Quotient (LQ)
data['LQ'] = (data['employment_sumsel'] / data['totalempsumsel']) / (data['employment_national'] / data['totalempnational'])
```

```python
data.head()
```

```python
# Calculate growth rates by sector and year
data['growth_sumsel'] = data.groupby('industry')['employment_sumsel'].pct_change()
data['growth_national'] = data.groupby('industry')['employment_national'].pct_change()
```

```python
data
```

```python
# Calculate average growth rates
average_growth_sumsel = data['growth_sumsel'].mean()
average_growth_national = data['growth_national'].mean()
```

```python
# Calculate DLQ
data['DLQ'] = (data['growth_sumsel'] / average_growth_sumsel) / (data['growth_national'] / average_growth_national)
```

```python
data
```

```python
# Calculate SLQ
data['SLQ'] = (data['LQ'] - 1) / (data['LQ'] + 1)
```

```python
data
```

```python
grouped_data1 = data[['year', 'industry', 'LQ', 'DLQ', 'SLQ']].sort_values(by=['industry', 'year'])
```

```python
grouped_data1
```

```python
output_path = "/content/groupeddata1.xlsx"

# export the dataframe to excel format, ensuring you use the correct variable for the dataframe
grouped_data1.to_excel(output_path, index=False, engine='openpyxl')
```

```python
# Plotting LQ, DLQ, and SLQ for each industry over time
metrics = ['LQ', 'DLQ', 'SLQ']
industries = grouped_data1['industry'].unique()

for metric in metrics:
    plt.figure(figsize=(10, 6))
    for industry in industries:
        industry_data = grouped_data1[grouped_data1['industry'] == industry]
        plt.plot(industry_data['year'], industry_data[metric], label=industry)

    plt.xlabel('Year')
    plt.ylabel(metric)
    plt.title(f'{metric} Over Time by Industry')
    plt.legend(title='Industry')
    plt.grid(True)
    plt.show()
```

## NGE, IME, and RCE

```python
# Calculate NGE, IME, and RCE
data['NGE'] = (data['employment_sumsel'] * average_growth_national).round(2)
data['IME'] = (data['employment_sumsel'] * (data['growth_national'] - average_growth_national)).round(2)
data['RCE'] = (data['employment_sumsel'] * (data['growth_sumsel'] - data['growth_national'])).round(2)
```

```python
data
```

```python
grouped_data2 = data[['year', 'industry', 'NGE', 'IME', 'RCE']].sort_values(by=['industry', 'year'])
```

```python
# Plotting NGE, IME, and RCE for each industry over time
metrics = ['NGE', 'IME', 'RCE']
industries = grouped_data1['industry'].unique()

for metric in metrics:
    plt.figure(figsize=(10, 6))
    for industry in industries:
        industry_data = grouped_data2[grouped_data2['industry'] == industry]
        plt.plot(industry_data['year'], industry_data[metric], label=industry)

    plt.xlabel('Year')
    plt.ylabel(metric)
    plt.title(f'{metric} Over Time by Industry')
    plt.legend(title='Industry')
    plt.grid(True)
    plt.show()
```

## DNGE, DIME, and DRCE

```python
# Calculate DNGE, DIME, and DRCE
data['DNGE'] = (data['employment_sumsel'].shift(1) * average_growth_national).round(2)
data['DIME'] = (data['employment_sumsel'].shift(1) * (data['growth_national'] - average_growth_national)).round(2)
data['DRCE'] = (data['employment_sumsel'].shift(1) * (data['growth_sumsel'] - data['growth_national'])).round(2)
```
